```python
def convert_sdf_to_xyz(sdf_file, output_dir):
    """Converts an SDF file with many molecules into separate XYZ files.

    Args:
      sdf_file: The path to the SDF file.
      output_dir: The directory where the XYZ files will be saved.
    """

    mols = pybel.readfile("sdf", sdf_file)

    for ind, mol in enumerate(mols):
        # print mol name
        name = mol.title
        # print(mol.title)
        xyz_file = os.path.join(output_dir, str(name) + ".xyz")
        mol.write("xyz", xyz_file)


if __name__ == "__main__":
    # make xyz folder
    root = "/home/santiagovargas/dev/qtaim_generator/data/qm9/"
    xyz = root + "xyz/"
    if not os.path.exists(xyz):
        os.mkdir(xyz)

    sdf_file = root + "gdb9.sdf"
    output_dir = root + "xyz/"
    convert_sdf_to_xyz(sdf_file, output_dir)
```

```python
from ase.atoms import Atoms
from ase.io import read, write
from xtb.ase.calculator import XTB
from ase.optimize.lbfgs import LBFGS
import rdkit
from rdkit import Chem
from rdkit.Chem import AllChem

import pandas as pd
import os
import numpy as np


def optimize_mol(
    smiles, mol_id, root="/home/santiagovargas/dev/qtaim_generator/data/tox21/xyz/"
):
    if not os.path.exists(root):
        os.mkdir(root)

    mol = Chem.MolFromSmiles(smiles)
    mol_H = Chem.AddHs(mol)
    AllChem.EmbedMolecule(mol_H)
    AllChem.UFFOptimizeMolecule(mol_H)
    # create folder from mol_id
    save_file = root + mol_id + ".xyz"
    # check if file exists
    if os.path.exists(save_file):
        return

    positions = mol_H.GetConformer().GetPositions()
    element_number = [atom.GetAtomicNum() for atom in mol_H.GetAtoms()]
    atoms = Atoms(numbers=element_number, positions=positions)
    atoms.calc = XTB(
        method="GFN2-xTB", solvent="None", accuracy=0.08, max_iterations=1000
    )
    opt = LBFGS(atoms, trajectory="./{}.traj".format(mol_id))
    opt.run(fmax=0.0001)
    traj_file = read("./{}.traj".format(mol_id), index=-1)
    write(filename=save_file, images=traj_file, format="xyz")


tox21_df = pd.read_csv("/home/santiagovargas/dev/qtaim_generator/data/tox21/tox21.csv")
root = "./xyz/"

for i in range(len(tox21_df)):
    # get a random row
    ind_grab = np.random.randint(0, len(tox21_df))

    row_test = tox21_df.iloc[ind_grab]
    smiles = row_test["smiles"]
    mol_id = row_test["mol_id"]
    try:
        optimize_mol(smiles, mol_id, root=root)
    except:
        print("error - {}".format(mol_id))
        continue
```

```python
# read libe
import pandas as pd
import os

df_libe = pd.read_json("../../data/libe/libe.json")
```

```python
def write_xyz(
    element,
    positions,
    mol_id,
    spin,
    charge,
    save_dir,
):
    with open(save_dir + mol_id + ".xyz", "w") as f:
        f.write(str(len(element)) + "\n")
        f.write("{} {}\n".format(charge, spin))
        for i, atom in enumerate(element):
            f.write(
                "{} \t\t {:.12f} \t\t {:.12f} \t\t {:.12f}\n".format(
                    atom, positions[i][0], positions[i][1], positions[i][2]
                )
            )


libe_save_dir = "../../data/libe/xyz/"
if not os.path.exists(libe_save_dir):
    os.mkdir(libe_save_dir)

for i, molecule_row in df_libe.iterrows():
    spin = molecule_row["spin_multiplicity"]
    charge = molecule_row["charge"]
    mol_id = molecule_row["molecule_id"]
    # print(molecule_row["molecule_graph"])
    element = [
        atom["name"] for atom in molecule_row["molecule_graph"]["molecule"]["sites"]
    ]
    positions = [
        atom["xyz"] for atom in molecule_row["molecule_graph"]["molecule"]["sites"]
    ]

    write_xyz(
        element=element,
        positions=positions,
        mol_id=mol_id,
        spin=spin,
        charge=charge,
        save_dir=libe_save_dir,
    )
```

```python
import numpy as np


def test(libe, libe_processed, n_tests=100):
    for i in range(n_tests):
        # get a random row
        ind_grab = np.random.randint(0, len(libe_processed))
        row_test = libe_processed.iloc[ind_grab]
        mol_id = row_test["names"].split(".")[0]
        test_spin = row_test["spin"]
        test_charge = row_test["charge"]

        # get the libe row with the same mol_id
        libe_row = libe[libe["molecule_id"] == mol_id]
        # print(libe_row)
        # print(row_test)
        libe_spin = libe_row["spin_multiplicity"].values[0]
        libe_charge = libe_row["charge"].values[0]
        print(
            "test_spin: ",
            test_spin,
            "libe_spin: ",
            libe_spin,
            "test_charge: ",
            test_charge,
            "libe_charge: ",
            libe_charge,
        )
        assert test_spin == libe_spin
        assert test_charge == libe_charge


test(libe=df_libe, libe_processed=libe_processed, n_tests=10)
```
